```python
# Import libraries
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# Import data
df = pd.read_csv('EngineeredRetail.csv', delimiter = ",") 
```

```python
## Explore the size of the data set
df.shape
```

```python
# Explore type of data and feature names
df.sample(5)
```

```python
# Assign the dataframe to the variable X
X = df.copy()
```

Draw a dendrogram using `method = "ward"` and matplotlib.

```python
# Find optimal number of clusters using a dendrogram
# YOUR CODE HERE
dendro = dendrogram(linkage(X, method = "ward"))
plt.title('Dendrogram')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
# Fit with three clusters
# YOUR CODE HERE
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')

y_hc=hc.fit_predict(X)
```

```python
# Set up dataframe with cluster labels
Xnew = X.copy()
Xnew["CLUSTERS"] = y_hc
Xnew.sample(10, random_state=0)
```

```python
# Count the number of observations in each cluster
Xnew['CLUSTERS'].value_counts()
```

```python
# Plot the three engineered variables using a pair-wise plot
# YOUR CODE HERE
var = ["Amount", "Frequency", "Recency"]
sns.pairplot(Xnew, vars=var, hue="CLUSTERS", palette=sns.color_palette("hls", 3))
```
